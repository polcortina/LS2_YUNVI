```python
import serial
import binascii
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

def twos_complement(hexstr):
    value = int(hexstr,16)
    if value & (1 << (7)):
        value -= 1 << 8
    return value

msp432 = serial.Serial('COM4', baudrate=115200, bytesize=8, parity=serial.PARITY_NONE, stopbits=1, timeout=10)  

n = 1000000

try:
    msp432.isOpen()
    print("Serial port is open")
except:
    print("Error")
    exit()

if(msp432.isOpen()):
    i = 0
    while (i<n):                
        msp432Serial = msp432.read()
        if (msp432Serial != ''):
            a= binascii.hexlify(msp432Serial)
            s = twos_complement(a)
            i = i + 1;
            print(s)
            
            
    msp432.close()
    
```
